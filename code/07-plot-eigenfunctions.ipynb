```python
# Load packages
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pickle

from FDApy import DenseFunctionalData, MultivariateFunctionalData
from FDApy.representation import DenseArgvals, DenseValues
from FDApy.preprocessing import MFPCA
from FDApy.visualization import plot, plot_multivariate

from nba import NbaScraper, ShotCharts

plt.rcParams.update({
    "text.usetex": True,
    "text.latex.preamble": r'\usepackage{amsfonts}'
})
```

```python
# Load MFPCA results
with open('./data/MFPCA.pickle', 'rb') as f:
    mfpca = pickle.load(f)
```

```python
pct = 100 * mfpca.eigenvalues / np.sum(mfpca.eigenvalues)
```

```python
# Mean
fig, ax = plt.subplots(
    2, 1,
    figsize=(4, 8),
    facecolor='white',
    edgecolor='black',
    dpi=300
)
ax[0] = ShotCharts.mean_chart(
    ax[0], mfpca, 0, 'Shots attempted'
)
ax[1] = ShotCharts.mean_chart(
    ax[1], mfpca, 1, 'Shots made',
)
ax[1].set_xlabel(f"Mean $\mu$", fontsize='x-large')

plt.savefig('./figures/mean.eps', transparent=True, format='eps')
plt.show()
```

```python
# Eigenfunctions
for idx in range(4):
    fig, ax = plt.subplots(
        2, 1,
        figsize=(4, 8),
        facecolor='white',
        edgecolor='black',
        dpi=300
    )
    ax[0] = ShotCharts.components_chart(
        ax[0], mfpca, idx, 0, 'Shots attempted'
    )
    ax[1] = ShotCharts.components_chart(
        ax[1], mfpca, idx, 1, 'Shots made'
    )
    ax[1].set_xlabel(f"Eigenfunction $\phi_{{{idx + 1}}}~({pct[idx]:.1f}\%)$", fontsize='x-large')

    plt.savefig(f'./figures/eigenfunction_{idx + 1}.eps', transparent=True, format='eps')
    plt.show()
```

```python
n_comp = mfpca.eigenfunctions.n_functional

fig, axes = plt.subplots(
    n_comp, 4 + 1,
    figsize=(4 * (4 + 1), 4 * n_comp),
    facecolor='white',
    edgecolor='black',
    dpi=300
)

axes[0, 0] = ShotCharts.mean_chart(
    axes[0, 0], mfpca, 0, 'Shots attempted'
)
axes[1, 0] = ShotCharts.mean_chart(
    axes[1, 0], mfpca, 1, 'Shots made',
)
axes[1, 0].set_xlabel(f"Mean $\mu$", fontsize='x-large')

for idx_o in np.arange(4):
    for idx_c in np.arange(n_comp):
        title = 'Shots attempted' if idx_c == 0 else 'Shots made'
        axes[idx_c, idx_o + 1] = ShotCharts.components_chart(
            axes[idx_c, idx_o + 1], mfpca, idx_o, idx_c, title
        )
    axes[1, idx_o + 1].set_xlabel(f"Eigenfunction $\phi_{{{idx_o + 1}}}~({pct[idx_o]:.1f}\%)$", fontsize='x-large')

#plt.savefig('./figures/eigenfunctions.eps', transparent=True, format='eps')
plt.show()
```

```python
class MidpointNormalize(mpl.colors.Normalize):
    def __init__(self, vmin=None, vmax=None, vcenter=None, clip=False):
        self.vcenter = vcenter
        super().__init__(vmin, vmax, clip)

    def __call__(self, value, clip=None):
        # I'm ignoring masked values and all kinds of edge cases to make a
        # simple example...
        # Note also that we must extrapolate beyond vmin/vmax
        x, y = [self.vmin, self.vcenter, self.vmax], [0, 0.5, 1.]
        return np.ma.masked_array(
            np.interp(value, x, y, left=-np.inf, right=np.inf)
        )

    def inverse(self, value):
        y, x = [self.vmin, self.vcenter, self.vmax], [0, 0.5, 1]
        return np.interp(value, x, y, left=-np.inf, right=np.inf)

fig, ax = plt.subplots(figsize=(8, 1), layout='constrained')

cmap = mpl.cm.seismic
norm = MidpointNormalize(vmin=-1, vcenter=0, vmax=1)

cb = fig.colorbar(
    mpl.cm.ScalarMappable(norm=norm, cmap=cmap),
    cax=ax, orientation='horizontal'
)
cb.set_label(label='Normalized value', fontsize='xx-large')

plt.savefig('./figures/colorbar2.eps', transparent=True, format='eps')
plt.show()
```
