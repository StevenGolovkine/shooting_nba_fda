```python
# Load packages
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pickle

from FDApy import DenseFunctionalData, MultivariateFunctionalData
from FDApy.representation import DenseArgvals, DenseValues
from FDApy.visualization import plot, plot_multivariate

from nba import NbaScraper, ShotCharts

plt.rcParams.update({
    "text.usetex": True,
    "text.latex.preamble": r'\usepackage{amsfonts}'
})
```

```python
# Load data
with open('./data/players_shots_density.pickle', 'rb') as f:
    players_shots_density = pickle.load(f)
with open('./data/players_shots_density_made.pickle', 'rb') as f:
    players_shots_density_made = pickle.load(f)
with open('./data/player_position.pickle', 'rb') as f:
    players_position = pickle.load(f)
```

```python
players_shots_density
```

```python
name = 'Stephen Curry'
fig, ax = plt.subplots(
    2, 1,
    figsize=(4, 8),
    facecolor='white',
    edgecolor='black',
    dpi=300
)
ax[0] = ShotCharts.shots_chart(
    ax[0], players_shots_density, name, 'Shots attempted', False
)
ax[1] = ShotCharts.shots_chart(
    ax[1], players_shots_density_made, name, 'Shots made', False
)
ax[1].set_xlabel(f"{name} - Guard", fontsize='x-large')

plt.savefig('./figures/curry_density.eps', transparent=True, format='eps')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 1), layout='constrained')

cmap = mpl.cm.Reds
norm = mpl.colors.Normalize(vmin=0, vmax=1)

cb = fig.colorbar(
    mpl.cm.ScalarMappable(norm=norm, cmap=cmap),
    cax=ax, orientation='horizontal'
)
cb.set_label(label='Normalized density', fontsize='xx-large')

plt.savefig('./figures/colorbar.eps', transparent=True, format='eps')
plt.show()
```
