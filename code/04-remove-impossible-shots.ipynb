```python
# Load packages
import json
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pickle
import requests

from nba_api.stats.endpoints import shotchartdetail
from nba_api.stats.static import players, teams

from nba import NbaScraper, ShotCharts
```

```python
# Load shots
with open('./data/player_shots_subset.pickle', 'rb') as f:
    players_shots = pickle.load(f)
```

```python
# The point (0, 0) is at the center of the basket.
# Add 52 to LOC_Y to move the (0, 0) to the center of the endline
players_shots.LOC_Y = players_shots.LOC_Y + 52
```

```python
players_shots[players_shots.LOC_Y < 0]
```

```python
# Remove impossible shots
X_COND = (-250 < players_shots.LOC_X) & (players_shots.LOC_X < 250)
Y_COND = (0 < players_shots.LOC_Y) & (players_shots.LOC_Y < 470)
players_shots_clean = players_shots[X_COND & Y_COND]
```

```python
f"Number of shots: {len(players_shots_clean)}"
```

```python
f"Number of shots: {len(players_shots_clean[players_shots_clean.SHOT_MADE_FLAG == 1])}"
```

```python
f"Percentage of made shots: {len(players_shots_clean[players_shots_clean.SHOT_MADE_FLAG == 1]) / len(players_shots_clean)}"
```

```python
# Save dataframe
with open('./data/players_shots_clean.pickle', 'wb') as f:
    pickle.dump(players_shots_clean, f, protocol=pickle.HIGHEST_PROTOCOL)
```
