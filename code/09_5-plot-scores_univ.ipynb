```python
# Load packages
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pickle
import scipy.stats as scs

from FDApy import DenseFunctionalData, MultivariateFunctionalData
from FDApy.representation import DenseArgvals, DenseValues
from FDApy.preprocessing import MFPCA, UFPCA
from FDApy.visualization import plot, plot_multivariate

from nba import NbaScraper, ShotCharts

from sklearn.preprocessing import StandardScaler

plt.rcParams.update({
    "text.usetex": True,
    "text.latex.preamble": r'\usepackage{amsfonts}'
})
```

```python
# Load data
with open('./data/players_shots_density.pickle', 'rb') as f:
    players_shots_density = pickle.load(f)
with open('./data/player_position.pickle', 'rb') as f:
    players_position = pickle.load(f)

# Load UFPCA results
with open('./data/UFPCA.pickle', 'rb') as f:
    ufpca = pickle.load(f)
with open('./data/scores_ufpca.pickle', 'rb') as f:
    scores = pickle.load(f)
with open('./data/UFPCA_reconstruction.pickle', 'rb') as f:
    fdata_reconstruction = pickle.load(f)
```

```python
# Reshape scores
scores = pd.DataFrame(scores)
scores.insert(loc=0, column='PLAYER_ID', value=players_shots_density.PLAYER_ID.values)
scores.insert(loc=0, column='PLAYER_NAME', value=players_shots_density.PLAYER_NAME.values)
scores = scores.join(players_position.set_index('PLAYER_ID'), how='left', on='PLAYER_ID', rsuffix='_')
```

```python
# Standardize the data
scaler = StandardScaler()
scaler.fit(scores[[0, 1, 2, 3]])
new_scores = scaler.transform(scores[[0, 1, 2, 3]])
```

```python
# First scores
COLORS = {
    'Guard': '#648FFF',
    'Forward-Guard': '#785EF0',
    'Forward': '#DC267F',
    'Forward-Center': '#FE6100',
    'Center': '#FFB000'
}

np.random.seed(1)
x = new_scores[:, 0]
y = np.repeat(0, len(scores)) + np.random.uniform(0, 3, size=len(scores))
c = scores['POSITION_'].map(COLORS)

fig, ax = plt.subplots(
    1, 1,
    figsize=(8, 1),
    facecolor='white',
    edgecolor='black',
    dpi=300
)

ax.vlines(x = 0, ymin=-10, ymax=10, linestyles='dashed', colors='#AAAAAA')
ax.scatter(x, y, c=c , s=5)
ax.scatter(
    x[[40, 153, 88, 49, 91]],
    y[[40, 153, 88, 49, 91]],
    c=c[[40, 153, 88, 49, 91]],
    edgecolors='black',
    s=25
)
ax.set_ylim(-10, 10)
ax.set_xlim(-5, 5)
ax.set_yticks([])

ax.annotate(  # Curry
    'Curry',
    xy=(x[40], y[40]), xycoords='data',
    xytext=(-3, -6), textcoords='data',
    bbox=dict(boxstyle="round", fc="w", linewidth=0.2),
    arrowprops=dict(arrowstyle="->", linewidth=0.5)
)
ax.annotate(  # Tatum
    'Tatum',
    xy=(x[153], y[153]), xycoords='data',
    xytext=(-0.25, 6), textcoords='data',
    bbox=dict(boxstyle="round", fc="w", linewidth=0.2),
    arrowprops=dict(arrowstyle="->", linewidth=0.5)
)
ax.annotate(  # James
    'James',
    xy=(x[88], y[88]), xycoords='data',
    xytext=(-0.5, -6), textcoords='data',
    bbox=dict(boxstyle="round", fc="w", linewidth=0.2),
    arrowprops=dict(arrowstyle="->", linewidth=0.5)
)
ax.annotate(  # Embiid
    'Embiid',
    xy=(x[49], y[49]), xycoords='data',
    xytext=(2, -6), textcoords='data',
    bbox=dict(boxstyle="round", fc="w", linewidth=0.2),
    arrowprops=dict(arrowstyle="->", linewidth=0.5)
)
ax.annotate(  # Jokic
    'Jokić',
    xy=(x[91], y[91]), xycoords='data',
    xytext=(2, 6), textcoords='data',
    bbox=dict(boxstyle="round", fc="w", linewidth=0.2),
    arrowprops=dict(arrowstyle="->", linewidth=0.5)
)
ax.set_title(f"Scores $c_1$", fontsize='x-large')

plt.savefig('./figures/univ_scores_1.eps', transparent=True, format='eps', bbox_inches='tight', pad_inches=0.5)
plt.plot()
```
